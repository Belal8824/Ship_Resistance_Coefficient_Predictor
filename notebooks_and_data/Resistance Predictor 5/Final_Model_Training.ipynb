Training Script

```python
# Importing the necessary libraries
import pandas as pd
from xgboost import XGBRegressor
```

```python
# To see the full columns
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

Loading the Dataset

```python
df = pd.read_csv('Series60_with_stratify_column.csv')
```

Exploratory Data Analysis (EDA)

```python
#Display the Dataset
df
```

```python
#Data statistics
df.describe()
```

```python
#Final Dataset Check
df.info()
```

Define Features and Target

```python
x_train = df[['CB', 'Cp', 'CW', 'Speed_Length_Ratio', 'Froude_Number', 'LCB_Final', 'S_Wetted_Coeff']]
y_train = df['Resistance_Coefficient_C']
```

Model Training

```python
model = XGBRegressor(n_estimators=290, learning_rate=0.05, n_jobs=6, max_depth=5, subsample=0.8, colsample_bytree=0.8)

#training of the model
model.fit(x_train, y_train)
```

Saving the Trained Model

```python
# Saving the trained model in separate file
model.save_model('ship_resistance_model.json')
```
